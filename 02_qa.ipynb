```python
from langgraph.graph import StateGraph, START,END
from langchain_groq import ChatGroq
from typing import TypedDict
from dotenv import load_dotenv
import os
```

```python
model = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
class LLMState(TypedDict):
    question:str
    ans:str
```

```python
def llm_res(state:LLMState)->LLMState:
    question=state["question"]
    prompt=f"You are an AI assistant that answers this {question}"
    ans=model.invoke(prompt)
    state["ans"]=ans.content
    return state
```

```python
state=StateGraph(LLMState)
state.add_node("llm_res",llm_res)
state.add_edge(START,"llm_res")
state.add_edge("llm_res",END)
```

```python
workflow=state.compile()
```

```python
workflow.invoke({"question":"What is the capital of India?","ans":""})
```
